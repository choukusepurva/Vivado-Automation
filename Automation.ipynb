```python
import os
import pandas as pd
import numpy as np
import matplotlib as matplot
```

```python
files = os.listdir("Modules")
files = [file for file in files if ".v" in file]
files
```

## Add Files

```python
os.system("vivado -mode batch -source tcl_add1.tcl")
```

## Find Top Modules

```python
top_modules = [file for file in files if "top" in file]
top_modules
```

## Create TOP Module Directory in Impl

```python
for mod in top_modules:
    os.mkdir("Impl/" + mod[:-2])
```

```python
for mod in top_modules:
    os.system("vivado -mode batch -source tcl_run.tcl -tclargs {}".format(mod))
    print("{} is Implmented and Reports are generated".format(mod))
```

```python
Signal = [0]*8
Logic = [0]*8
Static_Power = [0]*8
DynamicPower = [0]*8
Total_Power = [0]*8

DatapathDelay = [0]*8

F7_Muxes = [0]*8
LUT_as_Mem = [0]*8
LUT_as_Logic = [0]*8
Slice_LUTs = [0]*8
LUT2 = [0]*8
LUT3 = [0]*8
LUT4 = [0]*8
LUT5 = [0]*8
LUT6 = [0]*8
Slice_LUTs_distributed = [0]*8
LUTs_diff = [0]*8

i = 0

for mod in top_modules[:-1]:
    with open("Impl/{}/timing.txt".format(mod[:-2])) as f:
        lines = f.read()
        timing = lines.split("\n")
        timing_info = [line for line in timing if "Data Path Delay" in line]
        dpd = float(timing_info[0].split()[3][:-2])
        DatapathDelay[i] = dpd
        
    with open("Impl/{}/power.txt".format(mod[:-2])) as f:
        lines = f.read()
        power = lines.split("\n")
        power_info1 = [line for line in power if "Dynamic" in line]
        dpower = float(power_info1[0].split("|")[2].strip())
        DynamicPower[i] = dpower
        power_info2 = [line for line in power if "Static Power" in line]
        spower = float(power_info2[0].split("|")[2].strip())
        Static_Power[i] = spower
        power_info3 = [line for line in power if "Signals" in line]
        signal_power = float(power_info3[0].split("|")[2].strip())
        Signal[i] = signal_power
        power_info4 = [line for line in power if "Slice Logic" in line]
        logic_power = float(power_info4[0].split("|")[2].strip())
        Logic[i] = logic_power
        power_info5 = [line for line in power if "Total On-Chip Power (W)" in line]
        tpower = (power_info5[0].split("|")[2].strip())
        Total_Power[i] = tpower
        
    with open("Impl/{}/utilization.txt".format(mod[:-2])) as f:
        lines = f.read()
        uti = lines.split("\n")
        uti_info1 = [line for line in uti if "Slice LUTs" in line]
        util1 = int(uti_info1[0].split("|")[2].strip())
        Slice_LUTs[i] = util1
        
        
        uti_info2 = [line for line in uti if "F7 Muxes" in line]
        util2 = int(uti_info2[0].split("|")[2].strip())
        F7_Muxes[i] = util2
        
        uti_info3 = [line for line in uti if "LUT as Logic" in line]
        if (uti_info3):
            util3 = int(uti_info3[0].split("|")[2].strip())
            LUT_as_Logic[i] = util3

        uti_info4 = [line for line in uti if "LUT as Memory" in line]
        if (uti_info4):
            util4 = int(uti_info4[0].split("|")[2].strip())
            LUT_as_Mem[i] = util4
        
        uti_info6 = [line for line in uti if "LUT2" in line]
        if (uti_info6):
            util6 = int(uti_info6[0].split("|")[2].strip())
            LUT2[i] = util6
        
        uti_info7 = [line for line in uti if "LUT3" in line]
        if (uti_info7):
            util7 = int(uti_info7[0].split("|")[2].strip())
            LUT3[i] = util7
        
        uti_info8 = [line for line in uti if "LUT4" in line]
        if (uti_info8):
            util8 = int(uti_info8[0].split("|")[2].strip())
            LUT4[i] = util8
        
        uti_info9 = [line for line in uti if "LUT5" in line]
        if (uti_info9):
            util9 = int(uti_info9[0].split("|")[2].strip())
            LUT5[i] = util9
        
        uti_info10 = [line for line in uti if "LUT6" in line]
        if (uti_info10):
            util10 = int(uti_info10[0].split("|")[2].strip())
            LUT6[i] = util10
        

        
    i+=1
       
```

```python
result = np.array([Signal , Logic , Static_Power , DynamicPower , Total_Power , DatapathDelay , F7_Muxes , LUT_as_Mem , LUT_as_Logic , Slice_LUTs , Slice_LUTs_distributed , LUT2 , LUT3 , LUT4 , LUT5 , LUT6])
print(result)
```
